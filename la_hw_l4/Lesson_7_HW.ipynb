\end{cases}$$

$$\begin{pmatrix}
\left.\begin{matrix}

\end{matrix}\right|
\begin{matrix}

\end{matrix}
\end{pmatrix}.$$

$$det A =
\begin{pmatrix}
1 & -2 \\ 
3 & -4
\end{pmatrix} = -4+6=2$$
$$det A_1 =
\begin{pmatrix}
1 & -2 \\ 
7 & -4
\end{pmatrix} = -4+14=10$$
$$det A_2 =
\begin{pmatrix}
1 & 1 \\ 
3 & 7
\end{pmatrix} = 7-3=4$$

Корни СЛАУ:
$$x_1=\frac{detA_1}{detA}=\frac{10}{2}=5\\
x_2=\frac{detA_2}{detA}=\frac{4}{2}=2$$

Проверка:
$$\begin{cases}

\end{cases}$$
решение верное: 

$x_1=5\\

__4.__ Решить систему линейных уравнений методом Холецкого

$$\begin{cases}

__Решение__

Метод работает с квадратными симметричными матрицами $detA\neq0$ в этом случае корни СЛАУ можно вычислить по формуле $LL^{T}x=b=Ly$ 

Посчитаем коэффициенты матрицы $L$:  

$$l_{11}=\sqrt{a_{11}}=\sqrt{81}=9,$$
$$l_{21}=\frac{a_{21}}{l_{11}}=\frac{-45}{9}=5,$$
$$l_{31}=\frac{a_{31}}{l_{11}}=\frac{45}{9},$$

$$l_{22}=\sqrt{a_{22}-u_{21}^{2}}=\sqrt{50 - 25}=\sqrt{25}=5,$$
$$l_{32}=\frac{1}{l_{22}}\left ( a_{32}-l_{21}l_{31} \right)=\frac{1}{5}\left (-15-5\cdot(-5) \right)=2,$$

$$l_{33}=\sqrt{a_{33}-l_{13}^{2}-l_{23}^{2}}=\sqrt{38-5^2-2^2}=\sqrt{9}=3.$$

Составим матрицы $L$ и $L^{T}$: 

$$L=\begin{pmatrix}

\end{pmatrix}\,\,
L^{T}=\begin{pmatrix}

0 & 0 & 3
\end{pmatrix}$$

$Ly=b$ посчитаем $y$

```python
# поиск корней методом Крамера
# зададим матрицу и вектор свободных коэффициентов
A = np.array([[ 1,  1,  2,  4],
              [ 2,  5,  8,  9],
              [ 3, 18, 29, 18],
              [ 4, 22, 53, 33]])
b= np.array([[2,5,5,3]])

# провека условий использование метода
if np.linalg.det(A)==0 and len(A)!=len(A[0]):
    print("метод Крамера не подходит")

det_A = np.linalg.det(A)
b_t=b.transpose()
b_t
x=[0 for i in range(len(A))]
for i in range(len(A)):
    K=np.copy(A)
    for j in range(len(A)):
        K[j][i]=b_t[j]
    x[i]=np.linalg.det(K)/det_A

print('Корни СЛАУ:')
for i in range(len(x)):
    print(f'х{i+1}={x[i]:0.2f}')

```

```python
# поиск корней методом Гаусса
# зададим матрицу и вектор свободных коэффициентов
A = np.array([[ 1,  1,  2,  4],
              [ 2,  5,  8,  9],
              [ 3, 18, 29, 18],
              [ 4, 22, 53, 33]])
b= np.array([[2,5,5,3]])
A_=np.append(A, b.transpose(), axis=1)
# проверка на количество решений
rank_A = np.linalg.matrix_rank(A)
rank_A_ = np.linalg.matrix_rank(A_)
N = len(A[0])
if rank_A!=rank_A_:
    print("Матрица коэф. несовместная, неопределенная. СЛАУ не имеет решения")
else:
    if rank_A!=N:
        print("Матрица коэф. совместная, неопределенная. СЛАУ имеет бесконечное количество решений")     
    else:
        print("Матрица коэф. совместная, определенная. СЛАУ имеет единственное решение")

AA=np.zeros((len(A_),len(A_[0])))  
print(AA)
for i in range(len(A_)):
    for j in range(len(A_[0])):
        AA[i][j]=A_[i][j]/A_[i][i]
print(AA)
```
