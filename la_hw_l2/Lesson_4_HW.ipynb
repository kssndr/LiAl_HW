__1.__ Вычислить определитель:

   a)

$$\begin{vmatrix}
sinx & -cosx\\

$$Det

sinx & -cosx\\ 
cosx & sinx

__2.__ Определитель матрицы $A$ равен $4$. Найти:

   а) $det(A^{2})$;
    


для двух квадратных матриц справидливо $det(AB)=det(A)\cdot det(B)$
с другой стороны $\lambda \cdot A = \lambda \cdot det(A)$ 

Следовательно 
$$det(A^2)=det(A)\cdot det(A)=4\cdot4=16$$


```python
A =np.array([[1,1,1],[0,2,2],[3,3,0]])
print(f'Матрица А:\n{A}')
print(f'Матрица А^2:\n{A.dot(A)}')
print(f'Определитель А:\n{np.linalg.det(A):0.1f}')
print(f'Определитель А^2:\n{np.linalg.det(A.dot(A)):0.1f}')
```

   б) $det(A^{T})$;

$$det(A^{T})=det(A)$$

$$А = \begin{vmatrix}
1 & 1 & 1\\ 
0 & 2 & 2\\ 
3 & 3 & 0
\end{vmatrix}
А^{T} = \begin{vmatrix}
1 & 0 & 3\\ 
1 & 2 & 0\\ 
1 & 3 & 3
\end{vmatrix}$$
$$det(A)= \begin{vmatrix}
1 & 1 & 1\\ 
0 & 2 & 2\\ 
3 & 3 & 0
\end{vmatrix} = 1\cdot2\cdot0+1\cdot0\cdot3+1\cdot2\cdot3-1\cdot2\cdot3-1\cdot2\cdot3-1\cdot0\cdot0=6-6-6=-6$$
$$det(А^{T})= \begin{vmatrix}
1 & 0 & 3\\ 
1 & 2 & 3\\ 
1 & 3 & 0
\end{vmatrix} = 1\cdot2\cdot0+1\cdot3\cdot0+1\cdot3\cdot3-1\cdot2\cdot3-1\cdot3\cdot3-1\cdot0\cdot0=9-6-9=-6$$


```python
A =np.array([[1,1,1],[0,2,2],[3,3,0]])
print(f'Матрица А:\n{A}')
print(f'Транспонированная матрица А_Т:\n{A.T}')
print(f'Определитель А:\n{np.linalg.det(A):0.1f}')
print(f'Определитель А_Т:\n{np.linalg.det(A.T):0.1f}')
```

__3.__  Доказать, что матрица

$$\begin{pmatrix}
-2 & 7 & -3\\ 
4 & -14 & 6\\ 
-3 & 7 & 13
\end{pmatrix}$$
   
вырожденная.



Матрица называется _сингулярной,_ или _вырожденной,_ если ее определитель равен нулю.
Определитель равен нулю, если хотябы один столбец или строка линейнозависимы - нулевые, потому что система векторов, образующих матрицу будут лежать в пространстве, меньшем чем размерность пространства матрицы (схлопнется как минимум одно измерение)

В данной матрице мы можем обнулить вторую строку умножив вычтя из нее первую умноженную на -2.

```python
a = np.array([[-2,7,-3],[4,-14,6],[-3,7,13]])
print(f'Матрица a:\n{a}')
print(f'Определитель a:\n{np.linalg.det(a):0.1f}')
print(f'ранг матрицы а={np.linalg.matrix_rank(a)}')
```

__4.__ Найти ранг матрицы:

   а) $\begin{pmatrix}
1 & 2 & 3\\ 
1 & 1 & 1\\ 
2 & 3 & 4
\end{pmatrix};$



3 строка является суммой 1 и 2 - она станет нулевой если из нее вычесть первые две - соответственно ее можно отбросить и тем самым понизить ранг матрицы с 3 до 2

```python
a = np.array([[1,2,3],[1,1,1],[2,3,4]])
print(f'Матрица a:\n{a}')
print(f'Определитель a:\n{np.linalg.det(a):0.1f}')
print(f'ранг матрицы а={np.linalg.matrix_rank(a)}')
```

   б) $\begin{pmatrix}
0 & 0 & 2 & 1\\ 
0 & 0 & 2 & 2\\ 
0 & 0 & 4 & 3\\ 
2 & 3 & 5 & 6
\end{pmatrix};$

вычтем из 3 строки 1 и 2
$$\begin{pmatrix}
0 & 0 & 2 & 1\\ 
0 & 0 & 2 & 2\\  
2 & 3 & 5 & 6
\end{pmatrix}$$
Ранг матрицы — это порядок ее максимального невырожденного (или базисного) минора.
а тут мы видим что минор с дополнительной диагональю 3,2,1 - максимально большой минор для этой матрицы и не равен нулю $2*2*3-3*2*1=12-6=6$  
следовательно ранг матрицы равен 3

```python
a = np.array([[0,0,2,1],[0,0,2,2],[0,0,4,3],[2,3,5,6]])
d = np.array([[0,2,1],[0,2,2],[3,5,6]])
print(f'Матрица a:\n{a}')
print(f'Определитель минора:\n{np.linalg.det(d):0.1f}')
print(f'ранг матрицы а={np.linalg.matrix_rank(a)}')
```
