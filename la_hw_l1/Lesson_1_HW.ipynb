__5.__ Установить, является ли линейным подпространством:

а) совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;
    
б) все векторы, являющиеся линейными комбинациями данных векторов $\{u_{1}, u_{2}, ..., u_{n}\}$.

a) совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю, образовано тремя подпространствами: 
- $L_{1}$ множество всех векторов вида $(0,a,b)$
- $L_{2}$ множество всех векторов вида $(0,0,с)$
- $L_{3}$ множество всех векторов вида $(d,0,e)$ 

каждое подпространство само по себе является линейным - так как для каждого выполняется условие сложения элементов подпространства и умножения на скаляр: 
- $(0,a,b)+(0,x,y)=(0,a+x,b+y)$,     $\alpha \cdot (0,a,b) = (0,\alpha a,\alpha b)\in L_{1}$
- $(0,0,с)+(0,0,z)=(0,0,с+z)$,     $\gamma \cdot (0,0,с) = (0,0,\gamma с)\in L_{2}$
- $(d,0,e) +  (f,0,g) = (d+f,0,e+g)$,     $\delta \cdot (d,0,e) =(\delta d,0,\delta e)\in L_{3}$ 

, а вот для векторов из разны подпространст эти условия не выполняются - поэтому совокупность этих трех подпростраств не является линейным.

б) совокупность всех линейных комбинаций этих векторов образует некоторое подпространство исходного линейного пространства. Она называться _линейной оболочкой_ этого множества и обозначаться $L\{u_{1}, u_{2}, ... , u_{k}\}.$
